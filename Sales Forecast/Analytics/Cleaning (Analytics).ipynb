```python
df['Counsellor Name'].fillna(value='No Counsellor',inplace=True)
```

```python
df['Course Type'].fillna(value='Analytics',inplace=True)
df.isnull().sum()
```

```python
IsDigit=pd.DataFrame(df['Fees Received'].str.isdigit())
notDigit=df[IsDigit['Fees Received']==False]
notDigit
```

```python
df.drop(notDigit.index,inplace=True)
df
```

```python
df['Fees Received']=df['Fees Received'].astype('int64')
```

```python
df['Fees Total']=df['Fees Total'].astype('int64')
```

```python
df['FeesPending'] = df['Fees Total'] - df['Fees Received']
df
```

```python
df.drop(['Fees Pending'],axis=1,inplace=True)
df
```

```python
df['Month'] = pd.to_datetime(df['Month'])
```

```python
df.reset_index(drop=True,inplace=True)
```
